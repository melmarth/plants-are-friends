### Start with plant_soup2
that site has two tables that will be very useful but their fields aren't always in the same order. So we may want to build a dictionary

```python
plant_list2 = plant_soup2.find_all('a', {'href': re.compile(r'https://houseplantsexpert.com/.*')})[166:409]
plant_list2
```

```python
#Get the urls for the plants
href_list2 = [x['href'] for x in plant_list2]
href_list2
```

```python
#Get the names for the plants
name_list2 = [x.text for x in plant_list2]
name_list2
```

```python
# Make a dataframe
source2_df = pd.DataFrame({'common_name': name_list2, 'url': href_list2})
source2_df
```

```python
# Remove the rows for which the name is empty
source2_df = source2_df.loc[source2_df.common_name != '']

source2_df.head(50)
```

```python
# pull the fancy name into a different column and remove the duplicated rows
source2_df['fancy_name'] = source2_df['common_name']
source2_df['duplicate'] = False
for i in range(0, source2_df.shape[0]-1):
    next_row = i + 1
    if source2_df.iloc[next_row,1] == source2_df.iloc[i,1]:
        source2_df.iloc[i,2] = source2_df.iloc[next_row,0]
        source2_df.iloc[next_row,3] = True

source2_df['duplicate'].value_counts()
```

```python
source2_df = source2_df.loc[~source2_df.duplicate]
```

```python
source2_df = source2_df.drop(columns='duplicate')
```

```python
source2_df = source2_df.reset_index(drop=True)
source2_df
```
